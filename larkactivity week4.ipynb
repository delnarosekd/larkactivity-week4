```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv(r"C:\Users\DELNA ROSE\Downloads\Telco-Customer-Churn.xlsx - Telco-Customer-Churn.csv")
data.head()
```

# 1. Compare Churn count with respect to gender

```python
import seaborn as sns
sns.countplot(data['gender'], hue= data['Churn'])
plt.title("Churn count vs Gender",fontsize=20)
```

# 2. Find out how many female senior citizens there in the dataset

```python
female_senior=data[(data['gender']=='Female')& (data['SeniorCitizen']==1)]
female_senior
sns.countplot(x=female_senior['gender'])
plt.title("Female senior citizens",fontsize=20,fontweight='bold')
plt.figure(figsize=(10,20))
```

# 3. Compare Tenure with Total charges

```python
plt.figure(figsize=(8,8))
plt.scatter(data['tenure'],data['TotalCharges'],marker='*')
plt.title("Tenure vs Total charges",fontsize=20)
plt.xlabel("Tenure")
plt.ylabel("Totalcharges")
```

# 4. Find out which contract is preferred by the senior citizen

```python
senior=data[data['SeniorCitizen']==1]
senior
sns.countplot(x=senior['Contract'])
plt.title("Contract preferred by the seniorcitizens",fontsize=20)
```
